# 결측치

## pandas - isna(), isnull()
    결측치 원소를 True, 관측치를 False로 반환하는 메서드
    반대의 기능을 하는 메서드는 notna()와 notnull()
    
## pandas - fillna()
    결측치를 채워 넣기 위한 메서드
    value 인자에 결측치를 채워 넣을 값을 입력하면 딕셔너리 사용 가능
    method에 'bfill'은 뒤의 값을, 'ffill'은 앞의 값을 참고하여 결측 처리

## pandas - dropna()
    결측치가 있는 row 또는 column을 제거하기 위한 메서드
    how 인자에 'any'는 결측치가 하나라도 있는 경우, 'all'은 전체가 결측치인 경우 제거
    
## pandas - quantile()
    분위수를 연산하는 메서드로 이상치 필터링에 활용
    q 인자에 1개 이상의 확률값(0~1)을 입력하여 그에 맞는 분위수 산출 가능

```python
df_na2 = df.head(7)
df_na2
```

```python
df_na2.dropna()
```

```python
df_na2.dropna(how="all")
```

```python
df_na2.iloc[:,:-1]
```

```python
df_na2.iloc[:,:-1].dropna(how="all")
```

```python
print(df["Sepal_Length"].max())
print(df["Sepal_Length"].min())
print(df["Sepal_Length"].mean())
```

```python
print(df["Sepal_Length"].median())
print(df["Sepal_Length"].quantile())
```

```python
print(df["Sepal_Length"].quantile(q=0.25)) # 제 1 사분위수
print(df["Sepal_Length"].quantile(q=0.50)) # 제 2 사분위수
print(df["Sepal_Length"].quantile(q=0.75)) # 제 3 사분위수
```

문제 03. 평균을 기준으로 1.5 표준편차를 넘어서는 값을 이상치라고 간주할 때 Sepal.Length 변수를 기준으로 이상치인 row 개수는 몇 개인가?

```python
df = pd.read_csv("iris.csv")
df.head(2)
```

```python
sl_mean = df["Sepal.Length"].mean()
sl_std = df["Sepal.Length"].std()
print(sl_mean)
print(sl_std)
```

```python
cond_1 = df["Sepal.Length"] < (sl_mean - 1.5*sl_std)
cond_2 = df["Sepal.Length"] > (sl_mean + 1.5*sl_std)
print(cond_1)
print(cond_2)
```

```python
df_out = df.loc[(cond_1) | (cond_2),]
df_out
```

```python
df_out2 = df.loc[(cond_1) | (cond_2)]
df_out2
```
